```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
file = 'generalized_Babak_cosmo_030125_time_0.0hours.json.npy'
data = np.load(file,allow_pickle=True)
```

```python
print(data)
```

```python
positions = data['X']
time = data['tau']
```

```python
print(np.unique(positions))
```

```python
plt.imshow(positions[:,:,0],cmap = 'hot')
plt.gca().axis('off')
plt.title('CoSMO spatial distribution t=%s hours'%time[0],c='k')
#plt.figure(figsize=(20,20))
plt.show()
```

```python
plt.hist(positions[positions!=0].flatten(),color='k')
plt.ylabel('count')
plt.xlabel('strain type')
plt.title('strain distribution t=%s hours'%time[0],c='k')
plt.show()
```
