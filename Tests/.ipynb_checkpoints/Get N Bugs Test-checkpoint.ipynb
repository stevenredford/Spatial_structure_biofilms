```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
from scipy.signal import convolve2d
from scipy.signal import convolve
from mpl_toolkits.mplot3d import Axes3D
```

```python
side_len = 500 #microns
box_height = 2500 #microns
cell_size = 5 #microns
cell_to_diff_ratio = 10
density = 100 #mm^-2
n_bugs = 2
np.random.seed(2417)
```

```python
def sum_collapse_array(M,n):
    
    #from ChatGPT
    """
    Reduces a 3D tensor M by computing the sum of points in non-overlapping n x n x n blocks.
    
    Parameters:
        M (np.ndarray): The input 3D array.
        n (int): The size of the blocks used for density calculation.
        
    Returns:
        np.ndarray: The resulting 3D tensor P containing densities of each n x n x n block.
    """
    # Ensure M is a numpy array
    M = np.asarray(M)
    d1, d2, d3 = M.shape
    
    # Check if M dimensions are divisible by n
    if d1 % n != 0 or d2 % n != 0 or d3 % n != 0:
        raise ValueError("All dimensions of the tensor must be divisible by the block size n.")
    
    # Reshape M into a 6D array where each block is n x n x n
    reshaped = M.reshape(d1 // n, n, d2 // n, n, d3 // n, n)
    
    # Calculate density for each block
    densities = np.sum(reshaped != 0, axis=(1, 3, 5))# / (n * n * n)
    
    return densities
```

```python
def generate_X(side_len,cell_size,n_bugs,density):
    
    n_cells = int(density*cell_size**2*side_len**2/1e6)
    print(n_cells,'cells')
    
    #generate the matrix to contain bug positions
    cell_sides = int(side_len/cell_size)
    cell_height = int(box_height/cell_size)
    X = np.zeros((cell_sides,cell_sides,cell_height))
    
    #seed in a box of width side_len/4
    width = int(cell_sides/4)
    offset = int(cell_sides/2 - cell_sides/8)
    
    for nn in range(n_cells):
        x = int(width*np.random.rand()) + offset
        y = int(width*np.random.rand()) + offset
        z = 0 #int(width*np.random.rand())
        typ = np.floor((np.random.rand()-0.5)*n_bugs + 1)+1
        
        X[x,y,z] = typ
        
    return X
```

```python
def draw(M):
    
    X_1 = X.copy()
    X_2 = X.copy()
    X_1[X!=1] = 0
    X_2[X!=2] = 0
        
    dens1 = sum_collapse_array(X_1,cell_to_diff_ratio)
    dens2 = sum_collapse_array(X_2,cell_to_diff_ratio)
    
    """draw the concentrations"""
    fig, ax = plt.subplots(1,4,figsize=(50,25))
    ax[0].imshow(M[:,:,0], cmap='hot')
    ax[1].imshow(dens1[:,:,0], cmap='Greys')
    ax[2].imshow(dens2[:,:,0],cmap = 'Greys')
    ax[3].imshow(dens2[:,int(int(side_len/cell_size)/(2*cell_to_diff_ratio)),:], cmap='Greys')
    ax[0].set_title('original XY',fontsize=50)
    ax[1].set_title('density type 1 XY',fontsize=50)
    ax[2].set_title('density type 2 XY',fontsize=50)
    ax[3].set_title('density type 2 XZ',fontsize=50)
    ax[0].axis('off')
    ax[1].axis('off')
    ax[2].axis('off')
    ax[3].axis('off')
    
    return X_1, X_2
```

```python
X = generate_X(side_len,cell_size,n_bugs,density)
X_1, X_2 = draw(X)
```

```python
def expand_density_tensor(P, n):
    
    #from chatGPT
    """
    Expands a 3D tensor P by filling each n x n x n block with its corresponding density value,
    generating an array of the same size as the original tensor M.
    
    Parameters:
        P (np.ndarray): The input 3D tensor containing densities for each n x n x n block.
        n (int): The size of the blocks.
        
    Returns:
        np.ndarray: The resulting 3D array with the same size as the original tensor M.
    """
    # Ensure P is a numpy array
    P = np.asarray(P)
    
    # Get the shape of the expanded array
    d1, d2, d3 = P.shape
    expanded_shape = (d1 * n, d2 * n, d3 * n)
    
    # Create an array of the same size as the expanded tensor
    M_expanded = np.zeros(expanded_shape, dtype=P.dtype)
    
    # Fill each n x n x n block with the corresponding value from P
    for i in range(d1):
        for j in range(d2):
            for k in range(d3):
                M_expanded[
                    i * n:(i + 1) * n,
                    j * n:(j + 1) * n,
                    k * n:(k + 1) * n
                ] = P[i, j, k]
    
    return M_expanded
```

```python
#expand the density of 2 to the size of X
dens2 = expand_density_tensor(X_2, cell_to_diff_ratio)

#do some silly shit where we adjust X_1 by the density of X_2 in the region
X[X_1!=0] = X[X_1!=0]/dens2[X_1!=0]
```
