```python
positions = data.item()['X']
nutrients = data.item()['S']
time = data.item()['time']
```

```python
z = 5
plt.imshow(positions[:,:,0],cmap = 'hot_r')
plt.gca().axis('off')
plt.title('X-Y distribution t=%s hours'%time,c='k')
#plt.figure(figsize=(20,20))
plt.show()
```

```python
sz = np.shape(positions)
tr = transforms.Affine2D().rotate_deg(90)

plt.imshow(positions[int(sz[0]/2),:,:].transpose(),cmap = 'hot_r')
plt.gca().axis('off')
plt.title('Y-Z distribution t=%s hours'%time,c='k')
plt.gca().invert_yaxis()
#plt.figure(figsize=(20,20))
plt.show()
```

```python
plt.hist(positions[positions!=0].flatten(),color='k')
plt.ylabel('count')
plt.xlabel('strain type')
plt.title('strain distribution t=%s hours'%time,c='k')
plt.show()
```
